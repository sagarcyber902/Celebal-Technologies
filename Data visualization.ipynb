```python
import seaborn as sns
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target_names[iris.target]
```

```python
sns.displot(data=data, x="sepal length (cm)", kind="kde")  
```

```python
sns.displot(data=data, x="sepal width (cm)", kind="hist")
```

```python
sns.scatterplot(data=data, x="sepal length (cm)", y="petal length (cm)", hue="species")
```

```python
sns.boxplot(data=data, x="species", y="petal length (cm)")
```
